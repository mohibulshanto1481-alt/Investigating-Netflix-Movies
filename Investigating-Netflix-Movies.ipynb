```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
print(netflix_df)
```

```python
# Filter for movies released in the 1990s (assuming the 'release_year' column exists)
df_1990s = netflix_df[(netflix_df["release_year"] >= 1990) & (netflix_df["release_year"] < 2000)]
# Check the most frequent movie duration in the 1990s
most_frequent_duration = df_1990s['duration'].value_counts().idxmax()

# Save this as 'duration'
duration = most_frequent_duration
print(f"Most frequent movie duration in the 1990s: {duration}")
# Filter for action movies in the 1990s (assuming there's a 'genre' column)
action_movies_1990s = df_1990s[df_1990s['genre'] == 'Action']

# Count short action movies (duration < 90 minutes)
short_action_movies = action_movies_1990s[action_movies_1990s['duration'] < 90]

# Save the count as 'short_movie_count'
short_movie_count = short_action_movies.shape[0]
print(f"Number of short action movies in the 1990s: {short_movie_count}")
```
